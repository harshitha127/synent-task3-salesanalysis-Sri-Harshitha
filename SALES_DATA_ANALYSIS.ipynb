## **4. Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("SuperStoreOrders.csv")
df.head()
```

## **6. Understanding the Dataset**

```python
df.shape
df.columns
df.info()
```

## **7. Data Preprocessing**

**7.1 Fix column names if needed**

```python
df.columns
```

**7.2 Convert Date Column**

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True)
df['Month'] = df['order_date'].dt.to_period('M')
```

## **8. Sales Analysis**

**8.1 Monthly Revenue Trends**

```python
df['sales'].dtype
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['sales'].isnull().sum()

monthly_sales = df.groupby('Month')['sales'].sum()

monthly_sales.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

**8.2 Top-Selling Products**

```python
top_products = df.groupby('product_name')['sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=60)
plt.show()
```

**8.3 Profit Analysis**

```python
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
```

```python
sns.histplot(df['profit'])
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

## **9. Key Insights**



*   Monthly sales trends indicate fluctuations, suggesting possible seasonal demand patterns.
*   A small number of products contribute a large portion of total revenue, indicating key revenue drivers.

*   Profit distribution shows both positive and negative values, meaning some sales lead to losses.
*   High discounts or operational costs may be affecting profitability in certain transactions.

*   Focusing on high-performing products and reducing loss-making transactions can improve overall business performance.

## **10. Conclusion**

This analysis provides meaningful insights into sales trends, product performance, and profitability. These findings can help businesses make data-driven decisions to optimize revenue and improve efficiency.
